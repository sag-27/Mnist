#Data Loader Function in PYTORCH

```python
#relevant libraries
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import matplotlib.pyplot as plt
import numpy
```

```python
#transformation to load data into pytorch tensor and normalize it from 0 to 1(by default)
transform = transforms.Compose([transforms.ToTensor()])

#training dataset
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
#testing dataset
mnist_testset = datasets.MNIST(root='./data', train=False, download=True, transform=transforms.ToTensor())

#loading in batches
train_loader = torch.utils.data.DataLoader(mnist_trainset, batch_size=64, shuffle=False)
test_loader = torch.utils.data.DataLoader(mnist_testset, batch_size=64, shuffle=False)
```

```python
print('Trainset:', len(mnist_trainset))
print('Testset:', len(mnist_testset))
```

```python
#printing a tensor in training set
print(mnist_trainset[1])
```

```python
#visual rep of image
def plot_images(images):
    # Create a grid of images
    grid = torchvision.utils.make_grid(images, nrow=16)

    # Transpose the tensor to convert from (C, H, W) to (H, W, C) format
    grid = grid.numpy().transpose((1, 2, 0))

    # Plot the grid of images
    plt.figure(figsize=(16,16))
    plt.imshow(grid, cmap='gray')
    plt.axis('off')
    plt.show()

# Get a batch of images from the test loader
images, labels = next(iter(test_loader))

# Plot the images
plot_images(images)
```

The above image has 16 examples of from the training dataset with each image of 28x28 resolution. Each pixel can range from 0 to 255 depending upon the brightness. This end the loading and preprocessing part (preprocessing only required normalization)
